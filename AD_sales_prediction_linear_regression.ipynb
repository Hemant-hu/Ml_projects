```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('advertising.csv')
```

```python
df.head()
```

# Checking the linearity of data

```python
plt.scatter(x=df['TV'],y=df['Sales']) #linear
```

```python
plt.scatter(x=df['Radio'],y=df['Sales'])  #not linear
```

```python
plt.scatter(x=df['Newspaper'],y=df['Sales'])  #not linear
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Applying Linear regerresion


```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred
```

# Metrics r2_score

```python
from sklearn.metrics import r2_score,mean_absolute_error
mean_absolute_error(y_test,y_pred)     # get error
```

```python
r2_score(y_test,y_test)   # get accuracy  1.0=1*100=100 accuracy
```

# Predictive System


```python
df.iloc[0]
```

```python
def predict_sales(tv_budget,radio_budget,newspaper_budget):
    features=np.array([[tv_budget,radio_budget,newspaper_budget]])
    results=lr.predict(features).reshape(1,-1)
    return results[0]
```

```python
tv_budget=230.1
radio_budget=37.8
newspaper_budget=69.2
sales=predict_sales(tv_budget,radio_budget,newspaper_budget)
```
